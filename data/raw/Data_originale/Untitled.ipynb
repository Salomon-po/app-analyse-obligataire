```python
dataset = pd.concat([df_US, df_Al, df_Co, df_Maroc, df_Fr, df_GB, df_Ja], ignore_index=True)
dataset.head()
```

```python
dataset.tail()
```

```python
print(dataset.shape)
```

```python
dataset.info()
```

```python
dataset = dataset.drop("coef_indexation", axis=1)
```

```python
dataset["date_emission"] = dataset["date_emission"].fillna(dataset["date_emission"].mode()[0])
```

```python
dataset["isin"] = dataset["isin"].fillna("Inconu")
```

```python
dataset["Coupon"] = pd.to_numeric(dataset["Coupon"], errors="coerce")
dataset["ytm"] = pd.to_numeric(dataset["ytm"], errors="coerce")
dataset["prix"] = pd.to_numeric(dataset["prix"], errors="coerce")
```

```python
dataset["Coupon"] = dataset["Coupon"].fillna(dataset["Coupon"].mode()[0])
dataset["ytm"] = dataset["ytm"].fillna(dataset["ytm"].mode()[0])
dataset["prix"] = dataset["prix"].fillna(dataset["prix"].mode()[0])
```
